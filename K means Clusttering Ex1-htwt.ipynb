```python
import pandas as pd
```

```python
df=pd.read_csv("htwt.csv")
```

```python
x=df.iloc[:,[0,1]].values
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as  plt
import numpy as np
```

```python
abc=[]
#for loop to find the wcss for each number of clusters
for i in range(1,5):
    model=KMeans(n_clusters=i)
    model.fit(x)
    abc.append(model.inertia_)
#model.inertia_ is used to find the WCSS 

plt.plot(range(1,5),abc)
plt.title("the elbow method")
plt.xlabel("NUmber of Clusters")
plt.ylabel("WCSS-within cluster of squared sums")
```

```python
model=KMeans(n_clusters=3)
model.fit(x)
```

```python
y_model=model.predict(x)
```

```python
plt.scatter(x[y_model==0,0],x[y_model==0,1])
plt.scatter(x[y_model==1,0],x[y_model==1,1])
plt.scatter(x[y_model==2,0],x[y_model==2,1])
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=300,c='yellow')
```
